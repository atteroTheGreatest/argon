```python
def load_from_file(filename):
    xs = []
    ys = []
    with open(filename) as f:
        for line in f:
            x, y = line.split()
            xs.append(float(x))
            ys.append(float(y))
    return xs, ys
```

```python
xs, ys = load_from_file('melting/potential_To_100.0_tau_0.005_n_5100Ksim.dat')
plt.plot(xs, ys)
```

```python
import os
```

```python
def visualize_parameters(basedir, output_suffix):
    filenames = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'coords' not in x]

    for filename in filenames:
        xs, ys = load_from_file(filename)
        plt.plot(xs, ys)
        plt.title(filename.split('/')[1].split('_')[0])
        plt.show()
```

```python
def get_T_after_termalisation(basedir, output_suffix, termalization_threshold=100):
    t_filename = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'temperature' in x][0]
    _, ts = load_from_file(t_filename)

    return np.average(ts[termalization_threshold:])

def get_P_after_termalisation(basedir, output_suffix, termalization_threshold=100):
    p_filename = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'P_' in x][0]
    _, ps = load_from_file(p_filename)

    return np.average(ps[termalization_threshold:])
```

```python
import math
import numpy as np


def compute_p(T):
    L = 3.5
    v = 4 / 3.0 * math.pi * L ** 3
    v
    N = 343
    k = 0.00831
    p = 3 / 2 * N * k * T / v
    return p

def p_average_expected_to_temp(filename):
    termalization_threshold = 100
    average_p = get_P_after_termalisation('gas/', filename, termalization_threshold)
    T_after_termalization = get_T_after_termalisation('gas/', filename, termalization_threshold)

    print 'Temperature after termalization:', T_after_termalization
    p = compute_p(T_after_termalization)
    print('Average from simulation: %s vs Clapeyron: %s' % (average_p, p))
    return average_p, p, T_after_termalization

filenames = [
    '500Ksim_smaller_box.dat',
    '71000Ksim.dat',
    '71500Ksim.dat',
    '72000Ksim.dat',
]
for filename in filenames:
    p_average_expected_to_temp(filename)
```
