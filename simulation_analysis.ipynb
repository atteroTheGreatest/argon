# Analiza symulacji dynamiki molekularnej argonu

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
def load_from_file(filename):
    xs = []
    ys = []
    with open(filename) as f:
        for line in f:
            x, y = line.split()
            xs.append(float(x))
            ys.append(float(y))
    return xs, ys
```

## 1. Badania symulacji było badanie stabilności algorytmu

Przy dużych wartościach $\tau$ takich jak np. 0.1 algorytm jest bardzo niestabilny, atomy wpadają na siebie - co wywołuje działanie ogromnej siły, albo wychodzą poza pudełko.

Dla wartości $\tau$ 0.002 algorytm jest już całkowicie stabilny. Natomiast nawet dla kroku 0.05 daje w większości poprawne wyniki.

## 2. Badanie stabilności kryształu 

Najkorzystniejszą wartością potencjału $a$ było $ a \approx 0.38$, które dalej wykorzystujemy w symulacji.

```python
import os
```

```python
# small function to visualize parameters of simulation
def visualize_parameters(basedir, output_suffix):
    filenames = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'coords' not in x]

    for filename in filenames:
        xs, ys = load_from_file(filename)
        plt.plot(xs, ys)
        plt.title(filename.split('/')[1].split('_')[0])
        plt.show()
```

Czy kryształ jest stabilny? Jak wyglądają jego parametry w trakcie symulacji?

```python
visualize_parameters('cristal/', '0Ksim.dat');
```

Kryształ jest stabilny.

Poniżej jest kod do obliczenia średnich temperatur i ciśnień i porównanie wyników z prawem Clapeyrona.

```python
def get_T_after_termalisation(basedir, output_suffix, termalization_threshold=100, second_threshold=2000):
    t_filename = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'temperature' in x][0]
    _, ts = load_from_file(t_filename)

    return np.average(ts[termalization_threshold:second_threshold])

def get_P_after_termalisation(basedir, output_suffix, termalization_threshold=100, second_threshold=2000):
    p_filename = [basedir + x for x in os.listdir(basedir) if output_suffix in x and 'P_' in x][0]
    _, ps = load_from_file(p_filename)

    return np.average(ps[termalization_threshold:second_threshold])
```

```python
import math
import numpy as np


def compute_p(T):
    L = 3.5
    v = 4 / 3.0 * math.pi * L ** 3
    v
    N = 343
    k = 0.00831
    p = 3 / 2 * N * k * T / v
    return p

def p_average_expected_to_temp(filename, second_threshold=2000):
    termalization_threshold = 100
    average_p = get_P_after_termalisation('gas/', filename, termalization_threshold, second_threshold)
    T_after_termalization = get_T_after_termalisation('gas/', filename, termalization_threshold, second_threshold)

    print 'Temperature after termalization:', T_after_termalization
    p = compute_p(T_after_termalization)
    print('Average from simulation: %s vs Clapeyron: %s' % (average_p, p))
    return average_p, p, T_after_termalization

filenames = [
    '500Ksim_smaller_box.dat',
    '71500Ksim.dat',
    '72000Ksim.dat',
]
Ts = []
Ps = []
```

### Sprawdzenie prawa Clapeyrona

```python
for filename in filenames:
    p, _, T = p_average_expected_to_temp(filename)
    Ps.append(p)
    Ts.append(T)
    
p, _, T = p_average_expected_to_temp('71000Ksim.dat', 500)
Ps.append(p)
Ts.append(T)
```

```python
plt.scatter(Ts, Ps)
plt.title(r'$P(T)$')
plt.xlabel(r'$T [K]$')
plt.ylabel(r'$P$')
```
